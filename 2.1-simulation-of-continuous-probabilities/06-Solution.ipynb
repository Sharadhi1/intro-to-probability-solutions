## Before we run the simulation, let's think about this for a second

### You have a surface with parallel lines that are distance $d$ apart

![](images/6-1.png)

### You drop a needle of length $l$ onto the surface

![](images/6-2.png)

### Now, we know that if the distance from the center of the needle to the closest line is $x$, and the acute angle formed by the needle is $\theta$...

![](images/6-3.png)

### For simplicity, let's assume $l=1$ and $d=1$

### Then we know that if $x>\frac{1}{2}$, then the needle won't reach the line so the probability of crossing it is ZERO

### But since the distance between the lines is 1, the $x$ value can never be greater than $\frac{1}{2}$, i.e. the needle is aways in reach

### So, the needle crosses the line when $\frac{1}{2}sin\theta \geq x$

### $\theta$ can have a value in the interval $[0, \pi]$ i.e. between 0 and 180 degrees

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
x = np.arange(0, np.pi, step = 0.1)
y = 0.5*np.sin(x)
```

```python
plt.plot(x,y)
plt.axhline(0.5, ls = 'dashed', color = 'r')
```

# We can use the graph above to answer the following question: for fixed $\theta$, what are the possible values for $x$ where the needle will cross the line?

## In other words, if the angle is 45 degrees (or $\frac{\pi}{4}$), how close to the line do we need to be for it to cross?

```python
plt.plot(x,y)
plt.axvline(np.pi*0.25, ls = 'dashed', color = 'g')
plt.axhline(0.5, ls = 'dashed', color = 'r')
plt.axhline(0, ls = 'dashed', color = 'r')
```

## What the plot above is showing us: the green line between the lower red dotted line and the blue function represents values for $x$ where the needle will cross the line (when $\theta$ is fixed at 45 degrees)

## The green line between the blue function and the upper red dotted line represents the possible $x$ values where the needle won't cross

# Therefore, when $\theta = \frac{\pi}{4}$, the probability of the needle crossing the line is equal to:

# $P = \frac{\text{Distance from 0 to function}}{\text{Distance from 0 to function}+\text{Distance from function to 0.5}} = \frac{0.5\cdot sin(\frac{\pi}{4})}{0.5} = sin(\frac{\pi}{4})$

## Therefore, for angle $\theta$, the probability of the needle crossing the line is $sin(\theta)$

## If we integrate this across all possible values of $\theta$, we get the total probability of the needle crossing the line

# $\int^{\pi}_{0}sin(x) dx = -cos(\pi) - cos(0) = 2$
